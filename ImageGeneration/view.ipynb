```python
import numpy as np
import matplotlib.pyplot as plt
import os, sys
%matplotlib inline
```

```python
target_path = "./logs/1_rectified_flow/eval/ckpt_8/samples_1.npz"
```

```python
sample0 = np.load(target_path)
sample = sample0['samples']

rows, columns = 8, 8
fig = plt.figure(figsize=(10, 10))

for i in range(1, columns*rows+1):
    fig.add_subplot(rows, columns, i)
    plt.imshow(sample[i-1])
    plt.xticks([])
    plt.yticks([])
plt.show()
```
